# Training

```python
from simpletransformers.classification import ClassificationModel, ClassificationArgs
import pandas as pd

model_args = ClassificationArgs(overwrite_output_dir= True, num_train_epochs=3)

model = ClassificationModel( "bert", 'bert-base-cased', args=model_args)
# model = ClassificationModel( "bert", 'bert-base-uncased', args=model_args)
# model = ClassificationModel( "roberta", 'roberta-base', args=model_args)
# model = ClassificationModel( "distilbert", 'distilbert-base-cased', args=model_args)
# model = ClassificationModel( "bert", 'dccuchile/bert-base-spanish-wwm-cased', args=model_args)
# model = ClassificationModel( "bert", 'dccuchile/bert-base-spanish-wwm-uncased', args=model_args)
# model = ClassificationModel( "albert", 'albert-base-v2', args=model_args)
# model = ClassificationModel( "xlnet", 'xlbn-base-cased', args=model_args)
# model = ClassificationModel( "bert", 'bert-base-cased', args=model_args)

df = pd.read_csv("haha_2021_train.csv").rename(columns={'is_humor':'labels'})#read training data file

train_df = df[:22000] #Training on first 22000 Tweets. Feel free to shuffle.
eval_df = df[22000:]

model.train_model(train_df) #Start model training
```

# Evaluation

```python
result, model_outputs, wrong_predictions = model.eval_model(eval_df)
```
