# Training

```python
from simpletransformers.classification import ClassificationModel, ClassificationArgs
import pandas as pd
import torch

# Set "regression = True" since this is a regression task
model_args = ClassificationArgs(overwrite_output_dir= True, num_train_epochs=2, regression = True)

#Change the pretrained model for various experiments
model = ClassificationModel( "bert", 'dccuchile/bert-base-spanish-wwm-cased', args=model_args, num_labels = 1)

df = pd.read_csv("haha_2021_train.csv").rename(columns={'humor_rating':'labels'})

train_df = df[:8000] #Training on first 8000 Tweets. Feel free to shuffle.
eval_df = df[8000:] 
del train_df['is_humor']

train_df = train_df.dropna()
model.train_model(train_df)
```

# Evaluation

```python
result, model_outputs, wrong_predictions = model.eval_model(eval_df)
```
