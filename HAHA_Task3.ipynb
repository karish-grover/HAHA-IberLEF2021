```python
from simpletransformers.classification import ClassificationModel, ClassificationArgs
import pandas as pd

#task3.csv file has the data for task 3 from haha_2021_train.csv, which are not NaN. 
df = pd.read_csv("task3 .csv")
df = df.rename(columns={'humor_mechanism':'labels'})

encoding_labels = {'reference':0, 'exaggeration':1, 'analogy':2, 'irony':3, 'insults':4, 'absurd':5, 'parody':6,
       'embarrassment':7, 'unmasking':8, 'misunderstanding':9, 'wordplay':10,
       'stereotype': 11}

#Mapping classes to int
df['labels'] = df['labels'].map(encoding_labels)

model_args = ClassificationArgs(num_train_epochs=3, overwrite_output_dir= True)

#Multi class classification, 12 classes i.e. num_labels = 12
model = ClassificationModel('xlnet','xlnet-base-cased',num_labels=12, args=model_args) 
#Feel free to experiment with other pre trained models
```

# Training and Evaluation

```python
train_df = df[:-500]
eval_df = df[-500:]

# Train the model
model.train_model(train_df)

# Evaluate the model
result, model_outputs, wrong_predictions = model.eval_model(eval_df)
```
